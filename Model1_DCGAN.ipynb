# Discriminator

#### Weight Initialization

```python
# code for Discriminator
def weights_init(m):
    """
    Custom weights initialization for neural network layers.

    The weights are initialized with a normal distribution with mean 0 and standard deviation 0.02.
    Batch normalization layers have their weights initialized with mean 1 and standard deviation 0.02,
    while the bias is initialized to 0.

    Parameters:
    - m: A PyTorch module (layer) whose weights need to be initialized.
    """
    classname = m.__class__.__name__
    
    # Check if the layer is a convolutional layer
    if classname.find("Conv") != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    
    # Check if the layer is a batch normalization layer
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

#### Discriminator Convolution Block

```python
def disc_conv(in_c, out_c, ks=4, stride=2, padding=1, bn=True, out_layer=False):
    """
    Create a convolutional block for the discriminator with optional batch normalization and activation.

    Parameters:
    - in_c: Number of input channels.
    - out_c: Number of output channels.
    - ks: Kernel size for the convolutional layer.
    - stride: Stride for the convolutional layer.
    - padding: Padding for the convolutional layer.
    - bn: Whether to include batch normalization after the convolution.
    - out_layer: Whether this block is the final block of the discriminator.

    Returns:
    - A sequential block containing the specified layers.
    """
    layers = [nn.Conv2d(in_c, out_c, kernel_size=ks, stride=stride, padding=padding, bias=False)]
    
    # Optionally add a batch normalization layer
    if bn:
        layers.append(nn.BatchNorm2d(out_c))
    
    # Choose activation: Sigmoid for the output layer and LeakyReLU otherwise
    layers.append(nn.Sigmoid() if out_layer else nn.LeakyReLU(0.2, inplace=True))
    
    return nn.Sequential(*layers)
```

#### Constructing the Discriminator

```python
# Constructing the Discriminator (D) using the defined convolutional blocks.
# The Discriminator is a neural network that tries to classify input images as real or fake.
D = nn.Sequential(
    disc_conv(1, 32, bn=False),        # First convolutional block without batch normalization
    disc_conv(32, 64),                 # Second convolutional block
    disc_conv(64, 128, ks=3),          # Third convolutional block with a kernel size of 3
    disc_conv(128, 1, out_layer=True, bn=False, padding=0)  # Output block without batch normalization
)

# Displaying the constructed Discriminator model
D
```

# Generator

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.layers = nn.Sequential(*[
                                      self.conv_block(100, 128, padding=0),
                                      self.conv_block(128, 64, stride=2, ks=3),
                                      self.conv_block(64, 32, stride=2),
                                      self.conv_block(32, 1, stride=2, bn=False, out_layer=True)
        ])
    # Our input is 100 dimensional random noise

    @staticmethod
    def conv_block(in_c, out_c, out_layer=False, ks=4, stride=1, padding=1, bias=False, bn=True):
        l = [nn.ConvTranspose2d(in_c, out_c, ks, stride=stride, padding=padding, bias=bias)]
        if bn: l.append(nn.BatchNorm2d(out_c))
        if out_layer: l.append(nn.Tanh())
        else: l.append(nn.ReLU(True))
        return nn.Sequential(*l)

    def forward(self, x):
        return self.layers(x)
```

```python
G = Generator()
G
```

```python
fixed_noise = torch.rand((64, 100, 1, 1))
G.apply(weights_init)
D.apply(weights_init)
```

```python
# define loss mechanism
criterion = nn.BCELoss()
criterion = criterion.to(device)
```

```python
# define optimisation mechanism
optim_G = torch.optim.Adam(G.parameters(), lr=0.0002, betas=(0.5, 0.999))
optim_D = torch.optim.Adam(D.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

```python
D = D.to(device)
G = G.to(device)
```

```python
EPOCHS = 100
FAKE_LABEL = 0.0 # 0.0
REAL_LABEL = 1.0 # 1.0
```

## Training Loop

```python
G_losses = []
D_losses = []
D_update_frequency = 1
G_update_frequency = 2

for epoch in range(1, EPOCHS + 1):
    for i, (images, labels) in enumerate(trainloader):
        images = images.to(device)
        fake_labels = torch.full((images.size(0), ), FAKE_LABEL).to(device)
        real_labels = torch.full((images.size(0), ), REAL_LABEL).to(device)

        # Train Discriminator
        if i % D_update_frequency == 0:
            D.zero_grad()
            d_real = D(images).view(-1)
            d_loss_real = criterion(d_real, real_labels)
            d_loss_real.backward()

            noise = torch.randn((images.size(0), 100, 1, 1)).to(device)
            fake_images = G(noise)
            d_fake = D(fake_images.detach()).view(-1)
            d_loss_fake = criterion(d_fake, fake_labels)
            d_loss_fake.backward()

            d_loss = d_loss_real + d_loss_fake
            optim_D.step()

        # Train Generator
        for _ in range(G_update_frequency):
            G.zero_grad()
            noise = torch.randn((images.size(0), 100, 1, 1)).to(device)
            fake_images = G(noise)
            d_fake = D(fake_images).view(-1)
            g_loss = criterion(d_fake, real_labels)
            g_loss.backward()
            optim_G.step()
        G_losses.append(g_loss.item())
        D_losses.append(d_loss.item())

        if i % 100 == 0:
            print(f"Epoch [{epoch}/{EPOCHS}], Batch [{i}], LOSS_D: {d_loss}, LOSS_G: {g_loss}")
```

## Generator vs Discriminator Loss Plot

```python
plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(G_losses,label="G")
plt.plot(D_losses,label="D")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## FID Score Calculation

```python
inception_model = models.inception_v3(pretrained=True).cpu().eval()
```

```python
# Generate a batch of random noise vectors
noise = torch.randn(64, 100, 1, 1).cpu()

# Generate images using the trained generator
with torch.no_grad():
    generated_images = G(noise).cpu()
```

```python
def get_features(images, model):
    images = images.cpu().float()  # Ensure images are on CPU in float format
    model = model.cpu().float()    # Ensure model is on CPU in float format

    # Resize images to match the Inception's input shape
    images = F.interpolate(images, size=(299, 299), mode='bilinear', align_corners=False)
    
    # Replicate the grayscale image across 3 channels
    images = images.repeat(1, 3, 1, 1)
    
    # Get features from the Inception model
    with torch.no_grad():
        features = model(images).detach().cpu().numpy()
    return features
```

```python
# Assuming you have a dataloader named `dataloader`
data_iter = iter(trainloader)
real_images, _ = next(data_iter)
```

```python
real_features = get_features(real_images, inception_model)
generated_features = get_features(generated_images, inception_model)
```

```python
# Compute mean and covariance
mu1, sigma1 = real_features.mean(axis=0), np.cov(real_features, rowvar=False)
mu2, sigma2 = generated_features.mean(axis=0), np.cov(generated_features, rowvar=False)
```

```python
# Compute FID
ssdiff = np.sum((mu1 - mu2)**2.0)
covmean = sqrtm(sigma1.dot(sigma2))
if np.iscomplexobj(covmean):
    covmean = covmean.real
fid = ssdiff + np.trace(sigma1 + sigma2 - 2.0 * covmean)
print("FID score:", fid)
```
